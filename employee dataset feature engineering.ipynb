# Q3. FIXING THE GENDER COLUMN


 The Gender column contains inconsistent values:

   "Male", "male", "MALE", "M", "1"  →  should all become  "Male"

   "Female", "female", "FEMALE", "F", "0"  →  should all become  "Female"

   a) Print all unique values in Gender column before cleaning

   b) Write a function clean_gender(val) that standardises the value

   c) Apply it to create a new column  Gender_Clean

   d) Verify: print unique values of Gender_Clean

```python
df['Gender'].unique()
```

```python
import numpy as np
def clean_gender(val):
    if pd.isna(val):
        return np.nan
    val=str(val).strip().lower()

    if val in ['male','m','1']:
        return 'Male'
    elif val in ['female','f','0']:
        return 'Female'
    else:
        return np.nan
```

```python
df['Gender_clean']=df['Gender'].apply(clean_gender)
```

```python
df['Gender_clean'].unique()
```

# Q4. FIXING MIXED-FORMAT COLUMNS 

 a) SALARY column contains values like "45000 rupees", "45k", "not disclosed",
 
    empty strings.  Convert everything to a numeric float (NaN where impossible).
    
    Store result in  Salary_Clean.


```python
def clean_salary(val):
    if pd.isna(val):
        return np.nan
    val=str(val).strip().lower()
    if val in ['','not disclosed']:
        return np.nan
    val=val.replace('rupees','').replace('k','000').strip()
    try:
        return float(val)
    except:
        return np.nan
        
```

```python
df['Salary_Clean']=df['Salary'].apply(clean_salary)
df['Salary_Clean']
```

```python
df.shape
```

```python
df['Salary_Clean'].unique()
```

# b) AGE column has values like "35 years", 999, -5.
#    Convert to numeric, replace impossible ages (< 18 or > 80) with NaN.
#    Store result in  Age_Clean.


```python
df['Age'].unique()
```

```python
def clean_age(val):
    if pd.isna(val):
        return np.nan
    val=str(val).strip().lower()
    val=val.replace('years','')
    for a in val:
        if a<'18' or a>'80':
            return np.nan
    try:
        return int(val)
    except:
        return np.nan


df['Age_Clean']=df['Age'].apply(clean_age) 
df['Age_Clean']
```

# c) PERFORMANCE_SCORE column has values like "78%", 150, -1.
#    Convert to numeric and cap valid range to 0–100 (replace outliers with NaN).
#    Store result in  Score_Clean.


```python
def clean_score(val):
    if pd.isna(val):
        return np.nan
    
    val = str(val).replace('%', '').strip()
    
    try:
        score = float(val)
        if 0 <= score < 100:
            return score
        else:
            return np.nan
    except:
        return np.nan

df['Score_Clean'] = df['Performance_Score'].apply(clean_score)
df['Score_Clean']
```

```python
df['Score_Clean'].unique()
```

# d) EXPERIENCE_YEARS has values like "5 yrs", "fresher".
#    Convert "fresher" → 0, strip " yrs", convert to numeric.
#    Store result in  Experience_Clean.

```python
def clean_experience(val):
    if pd.isna(val):
        return np.nan
    
    val = str(val).replace('fresher', '0').replace('yrs','').strip()
    
    try:
        return float(val)
    except:
        return np.nan

df['experience_Clean'] = df['Experience_Years'].apply(clean_experience)
df['experience_Clean']
```

# Q17. OUTLIER DETECTION — IQR METHOD on Salary 
─────────────────────────────────────────────────────────────────────────────
 a) Calculate Q1, Q3, and IQR for Salary_Clean
 
 b) Define lower and upper bounds: Q1 − 1.5×IQR  and  Q3 + 1.5×IQR

 c) Print count of outliers below lower bound and above upper bound

 d) Print the actual outlier salary values

 e) TREAT outliers using CAPPING (Winsorization):

    replace values below lower with lower, above upper with upper
    Store result in  Salary_Capped

 f) Verify: print min and max of Salary_Capped

```python
Q1 = df["Salary_Clean"].quantile(0.25)
Q3 = df["Salary_Clean"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Below:", (df["Salary_Clean"] < lower).sum())
print("Above:", (df["Salary_Clean"] > upper).sum())

df["Salary_Capped"] = df["Salary_Clean"].clip(lower, upper)
```

```python
# e) Capping (Winsorization)
df['Salary_Capped'] = df['Salary_Clean'].clip(lower=lower, upper=upper)
print("Min:", df['Salary_Capped'].min(), "Max:", df['Salary_Capped'].max())
```

# Q18. OUTLIER DETECTION — Z-SCORE METHOD on Score_Clean 

 a) Calculate Z-score for every value in Score_Clean
#    Z = (x − mean) / std

 b) Flag rows where |Z| > 3 as outliers

 c) Print count of flagged outliers

 d) Remove (drop) outlier rows from df for Score_Clean

 e) Print shape before and after removal

```python
mean = df["Score_Clean"].mean()
std = df["Score_Clean"].std()

df["Z"] = (df["Score_Clean"] - mean) / std

print("Outliers:", (df["Z"].abs() > 3).sum())

shape_before = df.shape
df = df[df["Z"].abs() <= 3]
print("Before:", shape_before)
print("After:", df.shape)

df.drop("Z", axis=1, inplace=True)
```

# Q25. BONUS — FREQUENCY ENCODING on City 
 ─────────────────────────────────────────────────────────────────────────────

 Frequency encoding replaces each category with how many times it appears.

 This is useful when many unique categories exist.



 a) Starting from df_raw, clean the City column (strip + title-case)

 b) Compute frequency count of each city

 c) Create column  City_Freq  by mapping each city to its frequency count

 d) Normalise City_Freq to 0–1 using Min-Max scaling and store as  City_Freq_Norm

 e) What happens if two different cities have the same frequency?
    Is that a problem? (comment your answer)

```python
df3 = df_raw.copy()

df3["City"] = df3["City"].str.strip().str.title()

freq = df3["City"].value_counts()

df3["City_Freq"] = df3["City"].map(freq)

# Min-Max normalization
df3["City_Freq_Norm"] = (
    (df3["City_Freq"] - df3["City_Freq"].min()) /
    (df3["City_Freq"].max() - df3["City_Freq"].min())
)

print(df3[["City","City_Freq","City_Freq_Norm"]].head())
```

e) Problem with equal-frequency cities:
 If Mumbai (50 employees) and Pune (50 employees) have the same count, they get
 the same encoded value. The model CANNOT distinguish them — two completely
 different cities become identical features. This is a known weakness of
 frequency encoding. Solution: combine frequency encoding with another encoding
(e.g. target encoding) or use embeddings for high-cardinality categoricals.
